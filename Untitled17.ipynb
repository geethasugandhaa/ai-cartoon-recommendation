```python
import pyttsx3
import time

def speak(text):
    engine = pyttsx3.init()
    engine.setProperty('rate', 150)
    engine.say(text)
    engine.runAndWait()
    engine.stop()
    time.sleep(0.3)

# START
print("Welcome to Cartoon Recommender AI!")
speak("Welcome to Cartoon Recommender AI!")

print("Answer a few questions and I will suggest a cartoon!")
speak("Answer a few questions and I will suggest a cartoon!")

# AGE
speak("Enter your age")
age = input("Enter your age: ")   # ✅ shown beside input
speak(age + " entered")

# MOOD
speak("What is your mood? happy, sad, bored, or excited")
mood = input("Enter your mood(happy/sad/bored/excited): ")
speak(mood + " selected")

# GENRE
speak("What type do you like? funny, action, or adventure")
genre = input("Enter type(funny/action/adventure): ")
speak(genre + " selected")

# LOGIC
speak("AI Recommendation")

if mood.lower() == "sad":
    cartoon = "Mickey Mouse Clubhouse"
elif mood.lower() == "happy":
    cartoon = "SpongeBob SquarePants"
elif mood.lower() == "bored":
    cartoon = "Tom and Jerry"
elif mood.lower() == "excited":
    cartoon = "Ben 10"
else:
    cartoon = "Doraemon"

# OUTPUT
print("You should watch:", cartoon)
speak("You should watch " + cartoon)
```
